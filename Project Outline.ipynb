# Chirped Contra-DC Project Outline

This explains the intent and outline of the chirped contra-DC (CCDC) project. The goals of this project divide in a short-term and a longer-term goals:

- **Short-term goal:** Demonstrate that it is possible to design unlimited-bandwidth contra-DCs using chirp.

- **Long-term goal:** Demonstrate that such contra-DC are UV-litho friendly, can be widely wavelength- and bandwidth-tuned without bandwidth residue.

Hence, the first part would be about fabricating many devices and showing that arbitrary bandwidth values can be attained. If that works, non-uniform heating will be introduced in order to tune the devices.
